```python
#data train
data =pd.read_csv("../input/lomba-satria-data-2022/train_fktp.csv",header=None)

data.columns=['PSTV01','PSTV02','PSTV15','FKP02','FKP03','FKP04','FKP05','FKP06','FKP07','FKP08','FKP09',
'FKP10','FKP11','FKP12','FKP14','FKP14A','FKP15','FKP15A','FKP16','FKP17','FKP18','FKP19','FKP20',
'FKP21','FKP22','FKP13CLASS']
data['FKP13CLASS'].replace(['Belum_Sehat', 'Sehat'], [0, 1], inplace=True)
data.index=np.arange(1,len(data)+1)
data.drop(['PSTV01','PSTV02','PSTV15','FKP02','FKP03','FKP04','FKP06','FKP14A','FKP15','FKP15A','FKP17'],axis=1,inplace=True)
```

```python
for i in data['FKP16']:
    if i in range(11,22):
        data['FKP16'].replace([i],[1],inplace=True)
    elif i in range(31,37):
        data['FKP16'].replace([i],[1],inplace=True)
    elif i in range(51,54):
        data['FKP16'].replace([i],[1],inplace=True)
    elif i in range(61,66):
        data['FKP16'].replace([i],[1],inplace=True)
    elif i in range(71,77):
        data['FKP16'].replace([i],[1],inplace=True)
    elif i in range(81,83):
        data['FKP16'].replace([i],[1],inplace=True)
    elif i in range(91,95):
        data['FKP16'].replace([i],[1],inplace=True)
    elif i ==98:
        data['FKP16'].replace([i],[2],inplace=True)
```

```python
for i in data['FKP20']:
    if i in range(5,11):
        data['FKP20'].replace([i],[1],inplace=True)
    elif i in range(11,16):
        data['FKP20'].replace([i],[1],inplace=True)
    elif i in range(16,20):
        data['FKP20'].replace([i],[1],inplace=True)
    elif i in range(20,30):
        data['FKP20'].replace([i],[1],inplace=True)
    elif i in range(30,35):
        data['FKP20'].replace([i],[1],inplace=True)
    elif i in range(35,40):
        data['FKP20'].replace([i],[1],inplace=True)
        
data['FKP20'].replace([98,99],[2,3],inplace=True)
```

```python
for i in data['FKP05']:
    if i in range(11,22):
        data['FKP05'].replace([i],[1],inplace=True)
    elif i in range(31,37):
        data['FKP05'].replace([i],[2],inplace=True)
    elif i in range(51,54):
        data['FKP05'].replace([i],[3],inplace=True)
    elif i in range(61,66):
        data['FKP05'].replace([i],[4],inplace=True)
    elif i in range(71,77):
        data['FKP05'].replace([i],[5],inplace=True)
    elif i in range(81,83):
        data['FKP05'].replace([i],[6],inplace=True)
    elif i in range(91,95):
        data['FKP05'].replace([i],[7],inplace=True)
```

```python
data['FKP07'].replace([1,2,3,4,5,6,7,8],[1,1,1,1,1,1,1,1],inplace=True)
data['FKP07'].replace([9],[2],inplace=True)
```
